```python
import numpy as np
from sklearn.cluster import KMeans
import imutils
import cv2 
import os
from collections import Counter
from colormath.color_objects import sRGBColor, LabColor, XYZColor
from colormath.color_conversions import convert_color
from colormath.color_diff import delta_e_cie2000
from natsort import natsorted, ns
```

```python

# Creater a color palette for each keyframe in the folder.
# Colors are displayed with theirs proportions.

# k : Number of color clusters
# image_folder_path : path for movie keyframes


image_folder_path = 'midnight/'
image_folder = os.listdir(image_folder_path)

k = 10
count = 0
movie_palette = None
frame_palette = None

for image in image_folder:

    frame = cv2.imread(image_folder_path + image)
    height, width = frame.shape[:2]
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame,(width//4,height//4))
    frame = np.reshape(frame, (frame.shape[0] * frame.shape[1],3))
    
    clt = KMeans(n_clusters=k) 
    labels = clt.fit_predict(frame)
    label_counts = Counter(labels)

# pass through each cluster label in descending order and stack horizontally 
# in an image the color with its proportion on the frame    
    for i in range(len(label_counts)):
       
        if frame_palette is None:
            if((label_counts.most_common(len(label_counts))[i][1])//100 == 0):
                frame_palette = np.full((100,1,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
            else:
                frame_palette = np.full((100,(label_counts.most_common(len(label_counts))[i][1])//100,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
        else:
            if((label_counts.most_common(len(label_counts))[i][1])//100 == 0):
                color = np.full((100,1,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
            else:
                color = np.full(((100,(label_counts.most_common(len(label_counts))[i][1])//100,3)), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
            frame_palette = np.hstack((frame_palette, color))
            
    frame_palette = cv2.cvtColor(frame_palette, cv2.COLOR_RGB2BGR)
    cv2.imwrite('palettes_temp/' + image[:-4] + '_palette.jpg' , frame_palette)
    frame_palette = None 
    count += 1
    print(count)
```

```python
# Create dominant movie color palette

# image_folder_path : path for keyframes palettes

image_folder_path = 'thor_palettes/'
image_folder = os.listdir(image_folder_path)

count = 0
colors = None
dominant_movie_palette = None
dominant_movie_palette_noProp = None

# Gets one row of each palette for faster calculation
for image in image_folder:
    
    img_line = []
    frame = cv2.imread(image_folder_path + image)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    if(colors is None):
        
        for i in range(frame.shape[1]):
            img_line.append(frame[0][i])
        colors = np.asarray(img_line, dtype='uint8')
        
    else:
        
        for i in range(frame.shape[1]):
            img_line.append(frame[0][i])
            
        img_line = np.asarray(img_line, dtype='uint8')
        colors = np.concatenate((colors, img_line))
    count += 1
    print(count)
    


# KMeans cluster with predetermined colors in the init array   
clt = KMeans(n_clusters=init.shape[0], init=init, n_init=1) 
labels = clt.fit_predict(colors)
label_counts = Counter(labels)

# pass through each cluster label in descending order and stack horizontally 
# in two images the color with and without its proportion on the keyframe palette

for i in range(len(label_counts)):    
    if dominant_movie_palette is None:
        dominant_movie_palette = np.full((100,(label_counts.most_common(len(label_counts))[i][1])//100,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
        dominant_movie_palette_noProp = np.full((100,100,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
    else:
        dominant_movie_palette_aux = np.full((100,(label_counts.most_common(len(label_counts))[i][1])//100,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
        dominant_movie_palette = np.hstack((dominant_movie_palette, dominant_movie_palette_aux))
        dominant_movie_palette_noProp_aux = np.full((100,100,3), clt.cluster_centers_[label_counts.most_common(len(label_counts))[i][0]], dtype='uint8')
        dominant_movie_palette_noProp = np.hstack((dominant_movie_palette_noProp, dominant_movie_palette_noProp_aux))
        
dominant_movie_palette = cv2.cvtColor(dominant_movie_palette, cv2.COLOR_RGB2BGR) 
dominant_movie_palette_noProp = cv2.cvtColor(dominant_movie_palette_noProp, cv2.COLOR_RGB2BGR) 
cv2.imwrite('palettes_temp/movie_dominant_color_palette.jpg',dominant_movie_palette)
cv2.imwrite('palettes_temp/movie_dominant_color_palette_noProportion.jpg',dominant_movie_palette_noProp)  
```

```python
# Create a movie color time palette

# iamge_folder_path : path for keyframes palettes
image_folder_path = 'thor_palettes/'
image_folder = os.listdir(image_folder_path)
image_folder = natsorted(image_folder, key=lambda y: y.lower())   

time_palette = None

for image in image_folder:
    
    frame = cv2.imread(image_folder_path + image)
    frame = imutils.rotate_bound(frame, 90)
    frame = cv2.resize(frame, (5,600))
    
    if(time_palette is None):
        time_palette = frame
    else:
        time_palette = np.hstack((time_palette, frame))
cv2.imwrite('palettes_temp/time_palette_thor.jpg',time_palette)
```
